### OCR 

```python
from PIL import Image
import pytesseract, cv2, io
import numpy as np
import re
```

```python
# Leer imagen desde archivo binario
with open("./Data/INE.jpg", "rb") as f:
    image_data = f.read()
```

```python
image = Image.open(io.BytesIO(image_data))
img_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
```

```python
# 1. Convertir a escala de grises
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)

# 2. Reducir ruido (filtrado mediano o bilateral)
gray = cv2.medianBlur(gray, 3)

# 3. Mejorar contraste (histograma igualado)
gray = cv2.equalizeHist(gray)
gray = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)
```

```python
# OCR con Tesseract
try:
    text = pytesseract.image_to_string(gray, lang="spa")
except Exception:
    text = pytesseract.image_to_string(gray)

print("Texto detectado:\n",text)
```

```python
# Limpieza básica
text = re.sub(r'[^A-Z0-9\sÁÉÍÓÚÑ]', ' ', text)
text = text.replace('0', 'O').replace('1', 'I')
text = re.sub(r'\s{2,}', ' ', text).strip()

print(text)
```
